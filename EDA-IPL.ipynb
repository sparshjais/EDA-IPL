```python
plt.figure(figsize=(14, 3))
sns.countplot('season', data=match_df, palette="Set1")
plt.title("Number of Matches played in each IPL Season", fontsize=15)
plt.xticks(fontsize=12)
plt.xlabel("Season", fontsize=12)
plt.ylabel("Matches", fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(10, 3))
sns.countplot('winner', data=match_df, palette="Set1")
plt.title("Number of Matches Winning by each teams in each IPL season", fontsize=15)
plt.xticks(rotation=90,fontsize=12)
plt.xlabel("Season",fontsize=12)
plt.ylabel("Matches",fontsize=12)
plt.show()
```

```python
plt.figure(figsize = (20, 5))
sns.countplot('venue', data = match_df, palette = "Set1")
plt.title("Number of matches plays in Venue Or Stadium in each IPL Season", fontsize=12)
plt.xticks(rotation=90,fontsize=12)
plt.xlabel("Venue or Stadium",fontsize=12)
plt.ylabel("Matches",fontsize=12)
plt.show()
```

```python
plt.figure(figsize=(9, 3))
sns.countplot(match_df['player_of_match'].value_counts()[:20], data=match_df, palette="Set1")
plt.title("Number of Player of the Match",fontsize=12)
plt.xticks(rotation=90,fontsize=12)
plt.xlabel('Each Match',fontsize=12)
plt.ylabel('No. of the Player of the Match',fontsize=12)
plt.show()
```

```python
data_group = match_df.groupby('season')['team1', 'team2'].agg(sum)
print("The Team-1  VS The Team-2")
data_group
```

```python
# We Will plot pie chart on the basis of Toss decision
toss = match_df.toss_decision.value_counts()
labels = np.array(toss.index)
sizes = toss.values
colors = ['#8EB897', '#DD7596']
plt.figure(figsize=(9, 3))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=False, startangle=80)
plt.title('Toss decisions',fontsize=12)
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(20, 3))
sns.countplot('team1', hue='toss_decision', data=match_df, palette="Set1")
plt.title("Number of matches won by Toss-Team 1", fontsize=12)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel("Teams",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.legend(fontsize=2)
plt.show()
```

```python
plt.figure(figsize=(20, 3))
sns.countplot('team2', hue='toss_decision', data=match_df, palette="Set1")
plt.title("Number of matches won by Toss-Team 2", fontsize=12)
plt.xticks(rotation=90, fontsize=12)
plt.xlabel("Teams",fontsize=12)
plt.ylabel("Count",fontsize=12)
plt.legend(fontsize=2)
plt.show()
```

```python
final_matches = match_df.drop_duplicates(subset=['season'], keep='last')
final_matches[['season', 'winner']].reset_index(drop=True).sort_values('season')
```

```python
top_players = match_df.player_of_match.value_counts()[:20]
plt.figure(figsize=(10, 3))
plt.title("Top Players of Match Winners",fontsize=12)
plt.xticks(rotation=90,fontsize=12)
plt.xlabel('Players',fontsize=12)
plt.ylabel('Count',fontsize=12)
top_players.plot.bar()
sns.barplot(x=top_players.index, y=top_players, orient='v', palette='coolwarm');
plt.show()
```
